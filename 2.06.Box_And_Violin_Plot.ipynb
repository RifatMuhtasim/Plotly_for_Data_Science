# Load the dataset

````python
df = pd.read_csv("dataset.csv")
df.head()
````

````python
df['Value'] = pd.to_numeric(df['Value'])
mapping = {'HONEY, BEE COLONIES, AFFECTED BY DISEASE - INVENTORY, MEASURED IN PCT OF COLONIES':'Disease',
           'HONEY, BEE COLONIES, AFFECTED BY OTHER CAUSES - INVENTORY, MEASURED IN PCT OF COLONIES':'Other',
           'HONEY, BEE COLONIES, AFFECTED BY PESTICIDES - INVENTORY, MEASURED IN PCT OF COLONIES':'Pesticides',
           'HONEY, BEE COLONIES, AFFECTED BY PESTS ((EXCL VARROA MITES)) - INVENTORY, MEASURED IN PCT OF COLONIES':'Pests_excl_Varroa',
           'HONEY, BEE COLONIES, AFFECTED BY UNKNOWN CAUSES - INVENTORY, MEASURED IN PCT OF COLONIES':'Unknown',
           'HONEY, BEE COLONIES, AFFECTED BY VARROA MITES - INVENTORY, MEASURED IN PCT OF COLONIES':'Varroa_mites'}
df['Data Item'] = df['Data Item'].map(mapping)
df.rename(columns={'Data Item':'Affected by', 'Value':'Percent of Colonies Impacted'}, inplace=True)
````

# Basic Plot

````python
fig = px.violin(
    data_frame = df,
    x = "Affected by",
    y = "Percent of Colonies Impacted",
    title = "What is killing our Bees."
)
fig.show()
````

# Jitter Plot

````python
fig = px.violin(
    data_frame = df,
    x = "Affected by",
    y = "Percent of Colonies Impacted",
    title = "What is killing our Bees.",
    points = "all",
    box = True
)
fig.show()
````

# Violin Mode

````python
filter_df = df[(df['State'] == "NEW YORK") | (df['State'] == "WASHINGTON")]

fig = px.violin(
    data_frame = filter_df,
    x = "Affected by",
    y = "Percent of Colonies Impacted",
    title = "What is killing our Bees.",
    color = "State",
    violinmode = "overlay",
    # color_discrete_map = {"NEW YORK": "magenta", "NEW YORK": "blue"}
)
fig.show()
````

# Control Hover

````python
fig = px.violin(
    data_frame = df,
    x = "Affected by",
    y = "Percent of Colonies Impacted",
    title = "What is killing our Bees.",

    hover_name = "Year",
    hover_data = ['State'],
    custom_data = ['Program']
)
fig.show()
````

# Marks to Subplots

````python
#Create Rows

fig = px.violin(
    data_frame = filter_df,
    x = "Affected by",
    y = "Percent of Colonies Impacted",
    title = "What is killing our Bees.",
    facet_row = "State"
)
fig.show()
````

````python
# Create Columns

fig = px.violin(
    data_frame = filter_df,
    x = "Affected by",
    y = "Percent of Colonies Impacted",
    title = "What is killing our Bees.",

    facet_col = "Period",
    facet_col_wrap = 2
)
fig.show()
````

# Template

````python
fig = px.violin(
    data_frame = df,
    x = "Affected by",
    y = "Percent of Colonies Impacted",
    title = "What is killing our Bees.",
    labels = {"State": "STATE", "Percent of Colonies Impacted": "% of Impacted"},
    template = "presentation"
)
fig.show()
````

# Animation

````python
fig = px.violin(
    data_frame = df,
    x = "Affected by",
    y = "Percent of Colonies Impacted",
    title = "What is killing our Bees.",
    template = "plotly_dark",

    animation_frame = "Year",
    range_y = [-20, 100]
)
fig.show()
````
