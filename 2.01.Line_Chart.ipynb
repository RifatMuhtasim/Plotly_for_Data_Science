```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
```

# Load the dataset

## Stocks Dataset

```python
stocks_df = px.data.stocks()
stocks_df['year'] = pd.to_datetime(stocks_df['date']).dt.year
stocks_df.head()
```

```python
fig = px.line(stocks_df, x="date", y=["GOOG", "AAPL"], title="Yearly Google & Apple Stock Price")
fig.show()
```

```python
#Draw line plot with graph objects
trace_google = go.Scatter(
    x = stocks_df['date'],
    y = stocks_df['GOOG'],
    mode = "lines",
    name = "Google",
    line = dict(color='blue')
)

trace_amazon = go.Scatter(
    x = stocks_df['date'],
    y = stocks_df['AMZN'],
    mode = "lines",
    name = "Amazon",
    line = dict(color="red")
)

#Define the layout for the chart
layout = go.Layout(
    title = "Google Vs Apple Stock Price over time",
    xaxis = dict(title="Year", tickformat="%Y"),
    yaxis = dict(title="Stock Price (USD)"),
    showlegend = True
)

fig = go.Figure(data=[trace_google, trace_amazon], layout=layout)
fig.show()
```

## Gapminder Dataset

```python
#Gapminder dataset
gapminder_df = px.data.gapminder()
gapminder_df.head()
```

```python
continent_grp = gapminder_df.groupby(['continent', 'year'])['lifeExp'].mean().reset_index()
continent_grp.head()
```

```python
fig = px.line(continent_grp, x="year", y="lifeExp", color="continent", markers=True, title="Yearly Life Expectancy by continents")
fig.show()
```

```python
#Bangladesh life Expectancy over time
bangladesh_df = gapminder_df[gapminder_df['country'] == "Bangladesh"]
display(bangladesh_df.head())

trace = go.Scatter(
    x = bangladesh_df['year'],
    y = bangladesh_df['lifeExp'],
    mode = 'lines',
    line = dict(color="orange")
)

layout = go.Layout(
    title = "Bangladesh Life Expectancy over year",
    xaxis = dict(title="Year"),
    yaxis = dict(title="lifeExp")
)

fig = go.Figure(data=trace, layout=layout)
fig.show()
```
